```python
import axelrod as axl
```

```python
y = [x for x in axl.strategies]
y = sorted(y, key=lambda x: x.name)
```

```python
s=""
for player in y:
    extra_str = ""
    if player.classifier["stochastic"]:
        extra_str += "*"
        
    if player.classifier["long_run_time"]:
        extra_str += " (LRT)"
    
    if player.classifier["inspects_source"]:
        extra_str += " (ISo)"
    
    if player.classifier["manipulates_source"]:
        extra_str += " (MSo)"
    
    if player.classifier["manipulates_state"]:
        extra_str += " (St)"   
        
    add_str = "\item " + player.name + extra_str + "\n"
    add_str=add_str.replace("_","\_")
    s+= add_str
    
print(s)
```
